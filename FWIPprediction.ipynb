## Data preprocessing

### Subtask:
Handle missing values, encode categorical features, and scale numerical features as necessary.


**Reasoning**:
Check for missing values and prepare data for processing by separating features and target.



```python
print("Missing values per column:")
print(data.isnull().sum())

X = data.drop('FWI', axis=1)
y = data['FWI']

print("\nFeature set X head:")
display(X.head())
print("\nTarget variable y head:")
display(y.head())
```

**Reasoning**:
Identify categorical and numerical columns and apply preprocessing steps (one-hot encoding and scaling) to them, then combine the processed features.



```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

categorical_cols = ['Classes', 'Region']
numerical_cols = [col for col in X.columns if col not in categorical_cols]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

X_processed = preprocessor.fit_transform(X)

print("Processed features shape:", X_processed.shape)
```

## Feature engineering

### Subtask:
Create new features that could improve the model's performance.


**Reasoning**:
Create polynomial and interaction features and add them to the processed feature set.



```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Create Polynomial Features for numerical columns
poly = PolynomialFeatures(degree=2, include_bias=False)
X_numerical = data[numerical_cols]
X_poly = poly.fit_transform(X_numerical)
poly_feature_names = poly.get_feature_names_out(numerical_cols)

# Create Interaction Features (a simple example by multiplying relevant columns)
# Consider interactions that might be physically relevant or suggested by data exploration
# For this example, let's consider Temperature * RH and FFMC * ISI
X_interaction = pd.DataFrame({
    'Temp_x_RH': data['Temperature'] * data['RH'],
    'FFMC_x_ISI': data['FFMC'] * data['ISI']
})

# Combine processed features, polynomial features, and interaction features
# First, convert the processed features (from StandardScaler and OneHotEncoder) back to a DataFrame for easier concatenation
# Note: The ColumnTransformer output is a numpy array. We need the feature names to convert it back to a DataFrame correctly.
# However, getting the feature names from ColumnTransformer with OneHotEncoder and StandardScaler can be complex.
# A simpler approach for this step, given we have the original numerical and categorical columns,
# is to apply the transformations separately and then combine, or carefully append the new features.

# Let's re-apply transformations to numerical and categorical data separately and then add new features
scaler = StandardScaler()
X_scaled_numerical = scaler.fit_transform(data[numerical_cols])
X_scaled_numerical_df = pd.DataFrame(X_scaled_numerical, columns=numerical_cols)

encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded_categorical = encoder.fit_transform(data[categorical_cols])
X_encoded_categorical_df = pd.DataFrame(X_encoded_categorical.toarray(), columns=encoder.get_feature_names_out(categorical_cols))


# Add polynomial and interaction features to the combined DataFrame
# We will append the new features to the numerical part before scaling, or add them afterwards
# Let's add them after the initial scaling for simplicity in this example, but it might be better to include them
# in the ColumnTransformer or scale them appropriately if their scale is very different.

# For simplicity, let's just add the polynomial and interaction features (unscaled) to the processed data
# This might not be the optimal approach for scaling, but demonstrates adding features.
# A better approach would be to include these in the ColumnTransformer or scale them.

# Given the previous step produced X_processed using a ColumnTransformer,
# let's try to append the new features to this existing array.

# Convert new features to numpy arrays
X_poly_np = X_poly # Already numpy from PolynomialFeatures
X_interaction_np = X_interaction.to_numpy()

# Ensure the number of samples matches
if X_processed.shape[0] == X_poly_np.shape[0] and X_processed.shape[0] == X_interaction_np.shape[0]:
    X_processed_new = np.concatenate((X_processed, X_poly_np, X_interaction_np), axis=1)
    print("Shape of X_processed after adding polynomial and interaction features:", X_processed_new.shape)
else:
    print("Error: Number of samples in new features does not match X_processed.")
    X_processed_new = X_processed # Keep the original if shapes don't match


# Update X_processed to include the new features
X_processed = X_processed_new
```

## Model selection

### Subtask:
Choose appropriate machine learning models for the prediction task.


**Reasoning**:
Identify suitable regression models from scikit-learn based on the task and dataset size, considering linear, tree-based, and ensemble methods.



```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# List of chosen models
chosen_models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=42)
}

print("Chosen models for FWI prediction:")
for name, model in chosen_models.items():
    print(f"- {name}: {type(model).__name__}")

print("\nExplanation for choosing these models:")
print("- Linear Regression, Ridge, Lasso: Simple, interpretable linear models. Useful as baselines and can handle potential multicollinearity (Ridge/Lasso).")
print("- Decision Tree Regressor: Non-linear model that can capture complex relationships. Prone to overfitting but serves as a base for ensemble methods.")
print("- Random Forest Regressor: Ensemble of Decision Trees, reduces overfitting and generally provides better performance than a single tree.")
print("- Gradient Boosting Regressor: Another powerful ensemble method that builds trees sequentially, often achieving high accuracy.")
```

## Model training

### Subtask:
Train the selected regression models on the preprocessed and feature-engineered data.


**Reasoning**:
Train each of the chosen regression models using the preprocessed and feature-engineered data and the target variable.



```python
trained_models = {}

for name, model in chosen_models.items():
    print(f"Training {name}...")
    model.fit(X_processed, y)
    trained_models[name] = model
    print(f"Finished training {name}.")

print("\nAll chosen models have been trained.")
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained models using appropriate regression metrics such as Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared.


**Reasoning**:
Import the necessary evaluation metrics and calculate MAE, MSE, and R-squared for each trained model, storing the results in a dictionary.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

evaluation_results = {}

for name, model in trained_models.items():
    print(f"Evaluating {name}...")
    y_pred = model.predict(X_processed)

    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    evaluation_results[name] = {
        'MAE': mae,
        'MSE': mse,
        'R-squared': r2
    }
    print(f"Finished evaluating {name}.")

print("\nEvaluation Results:")
for name, metrics in evaluation_results.items():
    print(f"\n{name}:")
    print(f"  MAE: {metrics['MAE']:.4f}")
    print(f"  MSE: {metrics['MSE']:.4f}")
    print(f"  R-squared: {metrics['R-squared']:.4f}")
```

## Model tuning

### Subtask:
Fine-tune the hyperparameters of the best-performing model (Gradient Boosting Regressor based on evaluation results) using GridSearchCV to improve its performance.


**Reasoning**:
Fine-tune the Gradient Boosting Regressor using GridSearchCV to find the best hyperparameters.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2]
}

# Instantiate the Gradient Boosting Regressor model
gbr = GradientBoostingRegressor(random_state=42)

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid,
                           scoring='neg_mean_squared_error', cv=5, n_jobs=-1)

# Fit GridSearchCV to the data
grid_search.fit(X_processed, y)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Print the results
print("Best parameters found by GridSearchCV:")
print(best_params)
print("\nBest negative mean squared error found by GridSearchCV:")
print(best_score)
```

## Prediction

### Subtask:
Use the best performing model (Gradient Boosting Regressor with tuned hyperparameters) to make predictions on the processed data.


**Reasoning**:
Instantiate, train, and predict using the tuned Gradient Boosting Regressor model.



```python
from sklearn.ensemble import GradientBoostingRegressor

# 1. Instantiate a GradientBoostingRegressor object using the best_params
tuned_gbr = GradientBoostingRegressor(**best_params, random_state=42)

# 2. Train this model on the entire X_processed and y dataset
print("Training the tuned Gradient Boosting Regressor...")
tuned_gbr.fit(X_processed, y)
print("Finished training the tuned Gradient Boosting Regressor.")

# 3. Use the trained model to make predictions on the X_processed data
y_pred_tuned = tuned_gbr.predict(X_processed)

print("\nPredictions made using the tuned model.")
```
